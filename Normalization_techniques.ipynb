```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('employees.json')
df
```

```python
# put only numberical columns from the given dataset
df_modified = df[['age','years_of_experience','salary']]
```

```python
df_modified
```

```python
df_modified.plot(kind = 'bar')
plt.show()
```

# Maximum absolute scaling method.
#This technique works by , diving the each value of the column by its respective absolute maximum value.
# suppose , age is the column 
age  = age/age.abs().max()

```python
max_scaled = df_modified.copy()
for column in max_scaled.columns:
    max_scaled[column] = max_scaled[column]/max_scaled[column].abs().max()
max_scaled
```

```python
max_scaled.plot(kind = 'bar')
plt.show()
```

# Minimum Maximum scaling method
This technique scales the value by subtracting the minimum value of that column and dividing by range of that column
# Ex . age is the column
# age = (age - age.min()) /  (age.max()-age.min())

```python
scaled = df_modified.copy()
for column in scaled.columns:
    scaled[column] = (scaled[column]-scaled[column].min()) / (scaled[column].max() - scaled[column].min())
scaled
```

```python
scaled.plot(kind = 'bar')
plt.show()
```

```python
# Z-score scaling method.

z_scaled = df_modified.copy()
for column in z_scaled.columns:
     z_scaled = (z_scaled[column]-z_scaled[column].mean())/z_scaled[column].std()
z_scaled
```
