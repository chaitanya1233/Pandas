# For each department, find the top 2 highest-paid employees who have more been paid highest

```python

emp_hum_res_dep = df[(df['department']=='Human Resource')]
top_2_highest_paid_emp_from_hum_res = emp_hum_res_dep.nlargest(2,'salary')
emp_product_dep = df[(df['department']=='Product')]
top_2_highest_paid_emp_from_product_dep =emp_product_dep.nlargest(2,'salary')
```
